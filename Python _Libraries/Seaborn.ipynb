# import important libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as  sns
```

# importing dataset

```python
# Titanic dataset
titanic = pd.DataFrame(sns.load_dataset('titanic'))

# Iris dataset
iris = pd.DataFrame(sns.load_dataset('iris'))

# Tips dataset
tips = pd.DataFrame(sns.load_dataset('tips'))

# Flights dataset
flights = pd.DataFrame(sns.load_dataset('flights'))

# Students Performance dataset
students_performance = pd.read_csv('StudentsPerformance.csv')
```

# 1-Continues Variables

## 1.Histogram

```python
sns.histplot(tips['total_bill'], kde=True, bins=20)
plt.title('Histogram of Total Bill (Tips Dataset)')
plt.show()
#Visualize the distribution of total bill amounts in the Tips dataset.
```

## 2. KDE Plot

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
plt.title('KDE Plot of Total Bill by Gender (Tips Dataset)')
plt.show()
#Show the density distribution of total bill amounts by gender.
```

## 3-Scatter plot

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Scatter Plot of Sepal Dimensions (Iris Dataset)')
plt.show()
#Compare sepal length and width for different Iris species.
```

# 2-Categorical Variables

#  Count Plot

```python
sns.countplot(x='class', data=titanic)
plt.title('Count Plot of Passenger Class (Titanic Dataset)')
plt.show()
#Count the number of passengers in each class.
```

# Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot of Total Bill by Day (Tips Dataset)')
plt.show()
#Display the distribution of total bill amounts across days
```

# Violin Plot

```python
sns.violinplot(x='species', y='petal_length', data=iris)
plt.title('Violin Plot of Petal Length by Species (Iris Dataset)')
plt.show()
#Show the distribution of petal lengths for different Iris species
```

# Strip Plot

```python
sns.stripplot(x='day', y='tip', data=tips, jitter=True)
plt.title('Strip Plot of Tips by Day (Tips Dataset)')
plt.show()
#Display individual tip amounts across days with some jitter for visibility
```

# Swarm Plot

```python
sns.swarmplot(x='sex', y='age', data=titanic)
plt.title('Swarm Plot of Age by Gender (Titanic Dataset)')
plt.show()
#Visualize age distribution by gender in the Titanic dataset.
```

# --- Multivariate Analysis ---

# Pair Plot

```python
sns.pairplot(iris, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
#Visualize pairwise relationships between features
```

# Heatmap

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap='YlGnBu')
plt.title('Heatmap of Flights Dataset')
plt.show()
```
