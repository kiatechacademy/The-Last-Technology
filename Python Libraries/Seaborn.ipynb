## 1-displot==> histogram==> maghadire continues

```python
plt.figure(figsize=(8,3))
sns.displot(data=df,x="salary",color="red",edgecolor="blue")
plt.title("Histogram")
plt.xlabel("salary")

#plt.savefig(r"E:\my courses\courses\mine\ai\Pythons\data visualization\datasets\seaborn\figures\fig1.PNG")
```

## 2-rugplot

```python
sns.rugplot(data=df,x="salary",height=0.2)
```

## 3-kdeplot==> tozie

```python
sns.kdeplot(data=df,x="salary",label="Salary")
sns.kdeplot(data=df,x="sales",label="Sales")
plt.legend(loc=(1.1,0.5))
```
